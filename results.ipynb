# Results
Note: Make sure pickle file is ready on result path. Pickle file generate from Detector and Correction

### Import Library

```python
import os
import pickle
import pandas
```

### Functions

```python
def load_pickle(path: str):
  with open(path, "rb") as f:
    d = pickle.load(f)
  return d

def evaluation(name, d, col, task):
  p, r, f = d.get_data_cleaning_evaluation(col)[:3]
  return {
    "Dataset": name,
    "Task": task,
    "Precision": "{:.2f}".format(p), 
    "Recall": "{:.2f}".format(r), 
    "F1 Score": "{:.2f}".format(f)
  }
```

### Detection and Correction Results

```python
evaluation_df = pandas.DataFrame(columns=["Dataset", "Task", "Precision", "Recall", "F1 Score"])
extension = ".pkl"

# Raha Detection
task = "raha_detection"
result_path = "./results/detection/raha"
results = [f for f in os.listdir(result_path) if f.endswith(extension)]

for item in results:
  path = os.path.join(result_path, item)
  name = item.split(".")[0]
  d = load_pickle(path)
  col = d.detected_cells

  evaluation_df.loc[len(evaluation_df)] = evaluation(name, d, col, task)

# DAE Detection
task = "dae_detection"
result_path = "./results/detection/DAE"
results = [f for f in os.listdir(result_path) if f.endswith(extension)]

for item in results:
  path = os.path.join(result_path, item)
  name = item.split(".")[0]
  d = load_pickle(path)
  col = d.detected_cells

  evaluation_df.loc[len(evaluation_df)] = evaluation(name, d, col, task)

# Raha Correction using Baran
task = "raha_correction (baran)"
result_path = "./results/correction/raha"
results = [f for f in os.listdir(result_path) if f.endswith(extension)]

for item in results:
  path = os.path.join(result_path, item)
  name = item.split(".")[0]
  d = load_pickle(path)
  col = d.corrected_cells

  evaluation_df.loc[len(evaluation_df)] = evaluation(name, d, col, task)

# DAE Correction using Baran
task = "dae_correction (baran)"
result_path = "./results/correction/DAE"
results = [f for f in os.listdir(result_path) if f.endswith(extension)]

for item in results:
  path = os.path.join(result_path, item)
  name = item.split(".")[0]
  d = load_pickle(path)
  col = d.corrected_cells

  evaluation_df.loc[len(evaluation_df)] = evaluation(name, d, col, task)
  
display(evaluation_df.sort_values(by="Dataset").reset_index(drop=True))
```
